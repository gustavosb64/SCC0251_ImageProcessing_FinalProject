```python
#!pip install opencv-python
import numpy as np
import matplotlib.pyplot as plt
import imageio as iio
import cv2
```

```python
img = iio.imread("./imagens/clean/amaz20.jpg")

plt.figure(figsize = (10, 10))
plt.imshow(img)
plt.title("Imagem Original")
plt.axis("off")
plt.show()
```

```python
img_blue = img[:,:,0]
img_green = img[:,:,1]
img_red = img[:,:,2]

plt.subplot(1,3,1)
plt.title("Blue channel")
plt.imshow(img_blue, cmap = "gray")
plt.subplot(1,3,2)
plt.title("Green channel")
plt.imshow(img_green, cmap = "gray")
plt.subplot(1,3,3)
plt.title("Red channel")
plt.imshow(img_red, cmap = "gray")
plt.show()
```

```python
# Setting parameter values
t_lower = 0  # Lower Threshold
t_upper = 255  # Upper threshold

# Applying the Canny Edge filter
canny_img = cv2.Canny(img, t_lower, t_upper)

plt.rcParams['figure.figsize'] = [16, 8]

plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.subplot(1,2,2)
plt.imshow(canny_img, cmap = "gray")
plt.show()
```

```python
# Setting new parameter values
t_lower = 255  # Lower Threshold
t_upper = 255  # Upper threshold

# Applying the Canny Edge filter
canny_img = cv2.Canny(img, t_lower, t_upper)

plt.rcParams['figure.figsize'] = [16, 8]

plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.subplot(1,2,2)
plt.imshow(canny_img, cmap = "gray")
plt.show()
```

```python
canny_img_blue = cv2.Canny(img_blue, t_lower, t_upper)
canny_img_green = cv2.Canny(img_green, t_lower, t_upper)
canny_img_red = cv2.Canny(img_red, t_lower, t_upper)

plt.rcParams['figure.figsize'] = [16, 8]

plt.subplot(1,3,1)
plt.title("Blue channel")
plt.imshow(canny_img_blue, cmap = "gray")
plt.subplot(1,3,2)
plt.title("Green channel")
plt.imshow(canny_img_green, cmap = "gray")
plt.subplot(1,3,3)
plt.title("Red channel")
plt.imshow(canny_img_red, cmap = "gray")
plt.show()
```
